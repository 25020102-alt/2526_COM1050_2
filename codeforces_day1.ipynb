```python
#2180A
t_case = int(input())
 
for _ in range(t_case):
    l, a, b = map(int, input().split())
 
    visited = set()
    current = a
    visited.add(current)
 
    while True:
        current = (current + b) % l
        if current in visited:
            break
        visited.add(current)
 
    print(max(visited))
```

```python
import math

w, h, d = map(int, input().split())
n = int(input())

uoc_so = []
for i in range(1, int(math.isqrt(n)) + 1):
    if n % i == 0:
        uoc_so.append(i)         
    if i != n // i:           
        uoc_so.append(n // i) 
    
uoc_so.sort()

min_cuts = float('inf')
ans = None

for i in uoc_so:
    if w % i == 0:
        r = n // i

        for j in uoc_so:
            if j > r: break

            if r % j == 0 and h % j == 0:
                k = r // j

                if d % k == 0:
                    current = i + j + k - 3
                    if current < min_cuts:
                        min_cuts = current
                        ans = (i - 1, j - 1, k - 1)

if ans:
    print(f"{ans[0]} {ans[1]} {ans[2]}")
else:
    print(-1)
    
```

```python
import heapq
import sys

# Dùng sys.stdin.read để đọc input nhanh cho bài dữ liệu lớn
input = sys.stdin.readline

def dau(u, v):
    atk = -u[0]      
    def_val = -v[0]  
    
    if atk >= def_val:
        heapq.heappop(v) # Tương đương v.pop()
    else:
        new_def = def_val - atk
        heapq.heapreplace(v, -new_def)
        
s = int(input())
for _ in range(s):
    a, b = map(int, input().split())
    m = list(map(int, input().split()))
    n = list(map(int, input().split()))

    m = [-x for x in m]
    n = [-x for x in n]
    heapq.heapify(m)
    heapq.heapify(n)
            
    j = 0
    while m and n: 
        if j % 2 == 0: 
            dau(m, n)
        else: 
            dau(n, m)
        j += 1

    if not m: 
        print('Bob')
    else: 
        print('Alice')

   
```
