```python
a, n = list(map(int, input().split()))
m = input()
m = list(m)

for i in range(n):
    j = 0
    while j < a - 1:
        if m[j] == 'B' and m[j + 1] == 'G':
            m[j] = 'G'
            m[j + 1] = 'B'
            j += 2
        else:
            j += 1

print("".join(m))
            
```

```python
s = input()
s = list(s)
cnt = 0
for i in range(len(s) - 2):
    if s[i] == 'Q':
        for j in range(i + 1, len(s) - 1):
            if s[j] == 'A':
                for k in range(j + 1, len(s)):
                    if s[k] == 'Q': cnt += 1

print(cnt)
```
