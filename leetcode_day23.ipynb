```python
#PROBLEM 220
from bisect import *
from math import *

class Solution:
    def containsNearbyAlmostDuplicate(self, nums: List[int], indexDiff: int, valueDiff: int) -> bool:
        window = []

        for i, x in enumerate(nums):
            if i > indexDiff:
                j = nums[i - indexDiff - 1]
                window.pop(bisect_left(window, j))
        
            k = bisect_left(window, x - valueDiff)
            if k < len(window) and abs(window[k] - x) <= valueDiff:
                return True
            
            insort(window, x)
    
        return False
```

```python
#solution 1
from bisect import *

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        if len(nums) <= k:
            return [max(nums)]
        else:
            window = nums[: k]
            window.sort()
            t = [window[-1]]
            
            for i, x in enumerate(nums):
                if i >= k:
                    j = nums[i - k]
                    window.pop(bisect_left(window, j))     
                    insort(window, x)
                    t.append(window[-1])
            
            return t
```

```python
#solution 2
from collections import deque

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
     # Bước 1: Khởi tạo
     dq = deque() # Lưu trữ các CHỈ SỐ (index)
     res = []     # Mảng kết quả
    
     for i in range(len(nums)):
        # Bước 2: Kiểm tra phần tử quá hạn (Trượt ra khỏi cửa sổ bên trái)
        # i - k là chỉ số cuối cùng của cửa sổ trước đó
        if dq and dq[0] <= i - k:
            dq.popleft()
            
        # Bước 3: Loại bỏ các 'ứng cử viên' yếu hơn
        # Nếu số hiện tại lớn hơn số ở cuối deque, số ở cuối không bao giờ là max được nữa
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
            
        # Bước 4: Thêm chỉ số hiện tại vào Deque
        dq.append(i)
        
        # Bước 5: Ghi nhận kết quả (Khi cửa sổ đã đủ kích thước k)
        if i >= k - 1:
            res.append(nums[dq[0]])
            
     return res
```
